```python
facilities=pd.read_csv("Facilities_Database_20240823.csv")
facilities.fillna({'latitude':0, 'longitude':0},inplace=True)
display(facilities)
# facilities.drop(['latitude'], axis=1, inplace=True)
# facilities.drop(['longitude'], axis=1, inplace=True)
```

```python
subgroup=facilities.groupby('facsubgrp')
```

```python
mental_health=subgroup.get_group("MENTAL HEALTH")
display(mental_health)
```

```python
map_mental_health=folium.Map(location=[40.747525,-73.893776])
for index,row in mental_health.iterrows():
    lat=row["latitude"]
    lon=row["longitude"]
    name=row["facname"]
    newMarker=folium.Marker([lat,lon],popup=name)
    newMarker.add_to(map_mental_health)
map_mental_health.save(outfile='facilityLocations_mentalhealth.html')
```

```python
post=int(input("What is your postcode? "))
found=False
postcodes=mental_health.groupby("postcode")
if (mental_health == post).any().any():
  neighborhood=postcodes.get_group(post)
  display(neighborhood)
  mapNeighborhood=folium.Map(location=[40.747525,-73.893776])
  for index,row in neighborhood.iterrows():
    lat=row["latitude"]
    lon=row["longitude"]
    name=row["facname"]
    newMarker=folium.Marker([lat,lon],popup=name)
    newMarker.add_to(mapNeighborhood)
  mapNeighborhood.save(outfile='facilityLocations_neighborhoodmentalhealth.html')
```
